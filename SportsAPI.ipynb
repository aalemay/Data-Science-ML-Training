```python
import requests
import pandas as pd

url = "https://v1.american-football.api-sports.io/games?league=1&season=2021"
headers = {
  'x-rapidapi-key': 'd0a239f8ca062bdb08426d6e59de4798',
  'x-rapidapi-host': 'v1.american-football.api-sports.io'
}

response = requests.request("GET", url, headers=headers)
data = response.json()

# get game data
games_data = data.get("response", [])

# Convert data to DataFrame
game_list = []
for game in games_data:
    # Get score values safely with explicit conversion to integers
    home_score = int(game.get("scores", {}).get("home", {}).get("total", 0) or 0)
    away_score = int(game.get("scores", {}).get("away", {}).get("total", 0) or 0)
    game_dict = {
        "Home Team": game.get("teams", {}).get("home", {}).get("name", "Unknown"),
        "Away Team": game.get("teams", {}).get("away", {}).get("name", "Unknown"),
        "Home Score": home_score,
        "Away Score": away_score,
        "Date": game["game"]["date"]["date"],
        "Time": game["game"]["date"]["time"],
        "Time Zone": game["game"]["date"]["timezone"],
        "Home Win": 1 if home_score > away_score else 0
    }
    game_list.append(game_dict)

df_nfl = pd.DataFrame(game_list)

df_nfl
```
